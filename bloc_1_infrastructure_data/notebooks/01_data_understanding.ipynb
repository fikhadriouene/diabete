````python
df_raw = pd.read_sas("../data/LLCP2021.XPT")
print("dataset loaded successfully")
````

````python
nb_lines = df_raw.shape[0]
nb_columns = df_raw.shape[1]

print(f"lines number : {nb_lines}")
print(f"columns number : {nb_columns}")
````

````python
print(df_raw.info())

print("-"*50)
print("Force python to display columns informations")
print("-"*50)

# compute null columns values
valeurs_nulles = df_raw.isnull().sum()

# load columns data types
types_donnees = df_raw.dtypes

# 
df_analyse = pd.DataFrame({
    "type": types_donnees,
    "null_values": valeurs_nulles
}).reset_index().rename(columns={"index": "column_name"})

with pd.option_context("display.max_rows", None):
    display(df_analyse)
````

````python
cols_to_keep = [
    "DIABETE4",
    "_RFHYPE6",
    "_RFCHOL3",
    "_CHOLCH3",
    "_BMI5",
    "SMOKE100",
    "CVDSTRK3",
    "_MICHD",
    "_TOTINDA",
    "_FRTLT1A",
    "_VEGLT1A",
    "_RFDRHV7",
    "_HLTHPLN",
    "MEDCOST1",
    "GENHLTH",
    "MENTHLTH",
    "PHYSHLTH",
    "DIFFWALK",
    "SEXVAR",
    "_AGEG5YR",
    "EDUCA",
    "INCOME3",
    "CHCKDNY2",
    "CHECKUP1",
    "PERSDOC3"
]

# verifying existence cols 
missing_cols = [col for col in cols_to_keep if col not in df_raw.columns]

if missing_cols:
    raise ValueError(f"Colonnes manquantes : {missing_cols}")

# Creation of the dataframe with the selected cols
df_selected = df_raw[cols_to_keep].copy()

print("New shape with selected cols :", df_selected.shape)
````
